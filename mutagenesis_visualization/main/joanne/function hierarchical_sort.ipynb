```python
import numpy as np
import pandas as pd
from scipy.cluster import hierarchy
```

```python
#load dataset
dataset = pd.read_csv("../HRas166_RBD.csv", header = None)
```

```python
def hierarchical_sort(dataset):
    df_1 = pd.DataFrame(dataset)
    df_2 = df_1.fillna(0)
    Z = hierarchy.ward(df_2.T)
    g = hierarchy.leaves_list(Z)
    df_2sorted = df_2[g].copy()
    
    return df_2sorted
```

```python
test = hierarchical_sort(dataset)

print(test)
```
